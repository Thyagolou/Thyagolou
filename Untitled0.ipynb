```python
!pip install PyMuPDF

import fitz  # PyMuPDF

def pdf_para_string(caminho_do_pdf):
    # Abre o arquivo PDF
    pdf_documento = fitz.open ('/content/cmei.pdf')

    # Inicializa uma string para armazenar o texto
    texto_completo = ""

    # Itera pelas páginas do PDF
    for numero_pagina in range(pdf_documento.page_count):
        # Obtém a página atual
        pagina = pdf_documento[numero_pagina]

        # Extrai o texto da página e adiciona à string
        texto_completo += pagina.get_text("text")

    # Fecha o documento PDF
    pdf_documento.close()

    return texto_completo

# Substitua 'caminho/do/seu/arquivo.pdf' pelo caminho real do seu arquivo PDF
caminho_do_pdf = 'caminho/do/seu/arquivo.pdf'
texto_extraido = pdf_para_string(caminho_do_pdf)

# Imprime o texto extraído
print(texto_extraido)
```
