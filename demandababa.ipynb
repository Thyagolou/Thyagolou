```python
# Instalar as dependências necessárias no Colab
!pip install pandas PyPDF2 openpyxl

# Fazer o upload dos arquivos PDF
from google.colab import files
import os

# Faça o upload dos arquivos PDF
uploaded = files.upload()

# Crie uma pasta para armazenar os arquivos carregados
pdf_folder = '/content/pdf_files'
os.makedirs(pdf_folder, exist_ok=True)

# Mova os arquivos PDF para a pasta criada
for filename in uploaded.keys():
    os.rename(filename, os.path.join(pdf_folder, filename))

# Importar as bibliotecas necessárias
import re
import PyPDF2
import pandas as pd

# Lista para armazenar os dados extraídos
data_list = []

# Percorrer todos os arquivos na pasta
for filename in os.listdir(pdf_folder):
    if filename.endswith('.pdf'):
        pdf_path = os.path.join(pdf_folder, filename)

        # Abrir o arquivo PDF
        with open(pdf_path, 'rb') as pdf_file:
            pdf_reader = PyPDF2.PdfReader(pdf_file)

            # Percorrer todas as páginas do PDF
            for page_num in range(len(pdf_reader.pages)):
                page = pdf_reader.pages[page_num]
                page_text = page.extract_text()  # Extrair o texto da página

                # Definir padrões de regex para capturar todas as ocorrências
                nome_aluno_pattern = r'Nº INTENÇÃO:\s*\d+\s*(.*?)\s*NOME DA CRIANÇA:'
                data_nascimento_pattern = r'NOME DA CRIANÇA:\s*(.*?)\s*DATA NASCIMENTO:'
                mae_pattern = r'DATA NASCIMENTO:\s*(.*?)\s*NOME MÃE:'
                intencao_pattern = r'Nº INTENÇÃO:\s*(\d+)'
                modalidade_pattern = r'ESCOLA PARTICULAR:\s*(.*?)\s*MODALIDADE:'

                # Encontrar todas as ocorrências de cada informação
                nomes_alunos = re.findall(nome_aluno_pattern, page_text, re.DOTALL)
                datas_nascimento = re.findall(data_nascimento_pattern, page_text, re.DOTALL)
                maes = re.findall(mae_pattern, page_text, re.DOTALL)
                intencoes = re.findall(intencao_pattern, page_text)
                modalidades = re.findall(modalidade_pattern, page_text, re.DOTALL)

                # Garantir que todas as listas tenham o mesmo tamanho para evitar erros
                max_len = max(len(nomes_alunos), len(datas_nascimento), len(maes), len(intencoes), len(modalidades))

                for i in range(max_len):
                    NomeAluno = nomes_alunos[i] if i < len(nomes_alunos) else ''
                    DataNascimento = datas_nascimento[i] if i < len(datas_nascimento) else ''
                    mae = maes[i] if i < len(maes) else ''
                    intencao = intencoes[i] if i < len(intencoes) else ''
                    modalidade = modalidades[i] if i < len(modalidades) else ''

                    # Adicionar os dados extraídos à lista
                    data_list.append({
                        'arquivo': filename,
                        'página': page_num + 1,
                        'NOME COMPLETO DO ALUNO(A)': NomeAluno.strip(),
                        'DATA DE NASCIMENTO': DataNascimento.strip(),
                        'PAI/MÃE OU RESP. LEGAL': mae.strip(),
                        'N° INTENÇÃO (SISTEMA AZUL)': intencao.strip(),
                        'MODALIDADE': modalidade.strip()
                    })

# Converter a lista de dados para um DataFrame do pandas
data_frame = pd.DataFrame(data_list)

# Salvar os dados em um arquivo Excel
output_excel = '/content/dados_extraidos.xlsx'
data_frame.to_excel(output_excel, index=False, engine='openpyxl')

print(f'Dados extraídos salvos em {output_excel}')

# Baixar o arquivo Excel gerado
from google.colab import files
files.download(output_excel)
```
